# Evaluating if increase in limit balances will lead to lower default rates by conducting AB Testing. 

## The Problem: 

From our initial data analysis of credit card defaults, we saw a link between people with higher limit balances with lower default percentage rate. Does high default rate correlate with low limit balances & will providing higher limit balances to customers lead to a decrease in default rates? 

## The Potential Solution: 

Increasing limit balances may potentially show a decrease in limit balances based on our initial analysis. 

## The Method of Testing: 

To test this experiment, we will implement AB Testing by increasing default percentage on a sample size of customers of 1,500 customers by 20% and will compare them to a control group of 1,500 other customers whose limit balances will not increase. 

This experiment will be conducted and evaluated in a 3 month period where the default rate percentage will be monitored throughout 3 months in a monthly basis. 

The decision for this sample size was implemented by conducting a T-Test based on the current mean estimates for limit balances that we do have on customers. 

The variable of interest here is the default percentage compared between the test group (Group A) and the control group (group B). 

If default percentage in the test group decreases by over 10%, conclude that increasing limit balances among customers will lead to a decrease in default rates. This can be used to slowly roll out limit balance increase for other customers as well. If default rate does not decrease by up to 10% then observe for 3 more months before deciding on limit increase being a practice for existing customer groups. 

## Summary of Study:

- Group A: Test Group (Limit Balance will increase by 20%) 
- Group B: Control Group (Limit Balance stays the same) 
- Sample size (A): 1,500 
- Sample size (B): 1,500 
- Time Period: 3 months 

## Sampling test groups: 

For this experiment, we will take all users that have used their credit card at least 4 times in the past month as the starting group. 

This group will then be divided into sample sizes of 1,500 test group and 1,500 control group. We're using active users of credit card because users that use credit cards less than 4 times may not have enough activity to detect any changes in default rates. 

They will be randomly selected using a randomizer in Excel.

## Success Metric: 

The experiment will be considered a success if default percentage decreases by 10% at the end of the 3 month period for Group A and no significant changes takes place in the control group. This will be evaluated by doing T-Tests in a monthly period for both groups.  

## Secondary Metric: 

We will also see if the average bills among customers also increase as their limit balance are increased so, we will see if there is a correlation between increased bills vs. limit balances. This will help us measure if limit balances are actually being used by the test group which in turn may provide more reasons to increase limit balance (increase spending among customers as well while ensuring lower default rates)


## Roll - Out: 

The correspondence to customers will be sent as listed on the timeline above. After the initial email is sent, the test group's activity will be monitored (i.e. their spendings / their payments) in a monthly basis and be compared to the control group. These numbers will be evaluated in a monthly basis for each group. 

The test group will not recieve any correspondence regarding their limit increase except for the initial email before experiment begins. The control group will not recieve any correspondence at all. 

Each month's data will be checked for statistically relevance by using T-Tests. This will be done by comparing the starting point data with each month as well as comparing the changes between each month the experiment takes place as well. This will be done for both groups. 

At the end of the experiment, the groups will be evaluated to see if increasing limit balances at random led to decrease in default for the test group. In addition, the control group will be checked to see if the default rate stayed the same or didn't change significantlly (citing outside factors). 

If successful in showing decrease in default rates by 10% - the company will roll out a plan to increase default rates for another 10% of its customers by 20% (same as this experiment). This will allow customers to have flexibility and an increased spending budget and will allow increase revenues for the company based on interest rates payments as well as reduce its risk exposure if there is a link between increase limit balances and default rates.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
```

```python
x = cc_df['LIMIT_BAL']
sns.distplot(x)
plt.show()
```

```python
limit_to_default = cc_df[['LIMIT_BAL', 'default.payment.next.month']].copy()
limit_to_default['Range of Limits'] = (cc_df['LIMIT_BAL'] / 100000).round()
range_of_limits = (cc_df['LIMIT_BAL'] / 100000).round()
range_mean = limit_to_default.groupby('Range of Limits').mean()
range_mean
range_size = limit_to_default.groupby('Range of Limits').size()
```

```python
range_sd = limit_to_default.groupby('Range of Limits').std()
range_sd
```

```python
x = range_mean.index
y = range_mean['default.payment.next.month']
plt.scatter(x,y)
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x, p(x), "r--" )
plt.title('Limit Balances and Default Percentage Relationship')
plt.xlabel('Range of Limit Balances')
plt.ylabel('Percentage of Defaults')
plt.show()
```
